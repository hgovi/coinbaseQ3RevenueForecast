```python
import requests
import pandas as pd

# --- CoinGecko API (no key required) ---
url = "https://api.coingecko.com/api/v3/coins/coinbase-wrapped-staked-eth/market_chart"
params = {"vs_currency": "usd", "days": "365"}  # last 365 days
headers = {"User-Agent": "Mozilla/5.0"}

# Request data
r = requests.get(url, params=params, headers=headers)
r.raise_for_status()
res = r.json()

# --- Build DataFrame ---
df_price = pd.DataFrame(res["prices"], columns=["timestamp", "price_usd"])
df_cap   = pd.DataFrame(res["market_caps"], columns=["timestamp", "market_cap"])

# Merge and derive circulating (â‰ˆ total) supply
df = df_price.merge(df_cap, on="timestamp")
df["supply_cbeth"] = df["market_cap"] / df["price_usd"]

# Format timestamps
df["date"] = pd.to_datetime(df["timestamp"], unit="ms")

# --- Aggregate to quarterly averages ---
df_q = df.resample("Q", on="date").mean().reset_index()
df_q["Quarter"] = df_q["date"].dt.to_period("Q").astype(str)

# --- Display results ---
print(df_q[["Quarter", "price_usd", "supply_cbeth"]])
```

```python
import requests
import pandas as pd
import time
from datetime import datetime

# --- Define your desired range (1 year before the current 365-day window) ---
# Adjust if you're running this at a different date
start_date = datetime(2023, 10, 1)
end_date   = datetime(2024, 10, 1)

# Convert to UNIX timestamps (seconds)
start_unix = int(time.mktime(start_date.timetuple()))
end_unix   = int(time.mktime(end_date.timetuple()))

# --- CoinGecko API call ---
url = "https://api.coingecko.com/api/v3/coins/coinbase-wrapped-staked-eth/market_chart/range"
params = {
    "vs_currency": "usd",
    "from": start_unix,
    "to": end_unix
}
headers = {"User-Agent": "Mozilla/5.0"}

r = requests.get(url, params=params, headers=headers)
r.raise_for_status()
res = r.json()

# --- Build DataFrame ---
df_price = pd.DataFrame(res["prices"], columns=["timestamp", "price_usd"])
df_cap   = pd.DataFrame(res["market_caps"], columns=["timestamp", "market_cap"])

# Merge and derive supply
df = df_price.merge(df_cap, on="timestamp")
df["supply_cbeth"] = df["market_cap"] / df["price_usd"]
df["date"] = pd.to_datetime(df["timestamp"], unit="ms")

# --- Aggregate to quarterly averages ---
df_q = df.resample("Q", on="date").mean().reset_index()
df_q["Quarter"] = df_q["date"].dt.to_period("Q").astype(str)

# --- Display results ---
print(df_q[["Quarter", "price_usd", "supply_cbeth"]])
```

### Creating proxy for avg_fiat_balance using balance sheet data

```python
import pandas as pd
import numpy as np

# ------------------------------------------------------------
# 1. Quarterly balance sheet inputs (from 10-Q / 10-K)
# Values in thousands USD from filings
# ------------------------------------------------------------
data = pd.DataFrame({
    "Quarter": ["2023Q4","2024Q1", "2024Q2", "2024Q3", "2024Q4", "2025Q1", "2025Q2"],
    
    # Customer custodial funds (Assets)
    # e.g., from Balance Sheet Assets section
    "custodial_funds_asset": [4570845, 5201906, 4197837, 4035045, 6158949, 5369865, 5121640],
    
    # Customer custodial fund liabilities (for validation)
    "custodial_funds_liab": [4570845, 5201906, 4197837, 4035045, 6158949, 5369865, 5121640],
    
    # USDC (Assets) â€” from balance sheet line item
    "usdc_asset": [576028, 860669, 1056648, 871425, 1241808, 2225054, 2153824],
    
    # Optional diagnostics
    "cash_and_cash_equiv": [5139351, 6711400, 722535, 7723806, 8543903, 8051169, 7539388]
})

# ------------------------------------------------------------
# 2. Adjust for non-fiat portion (subtract USDC)
# ------------------------------------------------------------
data["adj_custodial_asset"] = data["custodial_funds_asset"] - data["usdc_asset"]

# ------------------------------------------------------------
# 3. Compute average fiat balance between quarters (in billions)
# ------------------------------------------------------------
data["avg_fiat_balance_bil"] = (
    (data["adj_custodial_asset"] + data["adj_custodial_asset"].shift(1)) / 2
) / 1e6  # convert thousands â†’ billions

# ------------------------------------------------------------
# 4. Optional: apply fiat-mix ratio if you want to scale by assumption (e.g., 80%)
# ------------------------------------------------------------
fiat_ratio = 0.8
data["avg_fiat_balance_bil_scaled"] = data["avg_fiat_balance_bil"] * fiat_ratio

# ------------------------------------------------------------
# 5. Display results
# ------------------------------------------------------------
pd.set_option("display.float_format", lambda x: f"{x:,.2f}")
print(data[["Quarter", "custodial_funds_asset", "usdc_asset",
            "adj_custodial_asset", "avg_fiat_balance_bil_scaled"]])
```

### loan_balance_index calculation

```python
import pandas as pd
import numpy as np

# ------------------------------------------------------------
# 1. Initialize quarterly periods
# ------------------------------------------------------------
quarters = ["2024Q1","2024Q2","2024Q3","2024Q4","2025Q1","2025Q2","2025Q3"]

_balance_df = pd.DataFrame({"Quarter": quarters})

# ------------------------------------------------------------
# 2. Input underlying data (replace placeholders with real metrics)
#    Units: USD billions for OI/Volume, USD for BTC price
# ------------------------------------------------------------
loan_balance_df["btc_open_interest_usd_bil"] = [30, 33.5, 37, 40, 42, 40.5, 48]
loan_balance_df["eth_open_interest_usd_bil"] = [8.5, 10.0, 12.5, 16.0, 19.0, 17.5, 21.0]
loan_balance_df["total_trading_volume_usd_bil"] = [240, 220, 210, 320, 300, 215, 290]
loan_balance_df["btc_price_usd"] = [27000, 29000, 31000, 40000, 48000, 52000, 55000]

# ------------------------------------------------------------
# 3. Normalize each input to Q1_2024 = 1.0
# ------------------------------------------------------------
def normalize(series):
    return series / series.iloc[0]

loan_balance_df["norm_open_interest"] = normalize(
    loan_balance_df["btc_open_interest_usd_bil"] + loan_balance_df["eth_open_interest_usd_bil"]
)
loan_balance_df["norm_volume"] = normalize(loan_balance_df["total_trading_volume_usd_bil"])
loan_balance_df["norm_btc_price"] = normalize(loan_balance_df["btc_price_usd"])

# ------------------------------------------------------------
# 4. Weighted composite index (weights: 40% OI, 40% Volume, 20% BTC Price)
# ------------------------------------------------------------
loan_balance_df["loan_balance_index"] = (
    0.4 * loan_balance_df["norm_open_interest"]
    + 0.4 * loan_balance_df["norm_volume"]
    + 0.2 * loan_balance_df["norm_btc_price"]
)

# ------------------------------------------------------------
# 5. Optional: Scale to mean = 1 for easier interpretation
# ------------------------------------------------------------
loan_balance_df["loan_balance_index"] /= loan_balance_df["loan_balance_index"].mean()

# ------------------------------------------------------------
# 6. Display summary
# ------------------------------------------------------------
pd.set_option("display.float_format", lambda x: f"{x:,.2f}")
print(loan_balance_df[[
    "Quarter",
    "loan_balance_index",
    "norm_open_interest",
    "norm_volume",
    "norm_btc_price"
]])
```
